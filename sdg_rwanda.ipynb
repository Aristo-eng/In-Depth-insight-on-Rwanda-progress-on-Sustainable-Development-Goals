```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plots settings
%matplotlib inline
sns.set(style="darkgrid")
sns.set(rc={'figure.figsize':(10,7)})
```

$ \color{blue}{\textbf{ GOAL 1: No Poverty }}$

$ \textbf{1.4.1 ::} $
Proportion of population living in households with access to basic services

```python
no_Poverty_1_4_1 = pd.read_csv('all_indicators/1-4-1.csv')
Observation_level = pd.pivot_table( no_Poverty_1_4_1, index = 'Year', columns = 'Observation-level footnotes', values = 'Value')
# The observation level categorized per year
Observation_level
```

```python
# ploting
Observation_level.plot(kind='bar')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
plt.title('Frequency Distribution of access to basic services - Rwanda') 
plt.xlabel('Year') 
plt.ylabel('Value')
```

$ \color{blue}{ \textbf{ GOAL 8: Decent Work and Economic Growth}} $

$ \textbf{8.1.1 ::} $
Annual growth rate of real GDP per capita $\newline$
$ \textbf{8.2.1 ::} $
Annual growth rate of real GDP per employed person $\newline$
$ \textbf{8.5.1 ::} $
Proportion of population using (a) safely managed sanitation services and (b) a hand-washing facility with soap and water $\newline$ $\newline$
$ \textbf{8.5.2 ::} $
Unemployment rate, by sex, age and persons with disabilities $\newline$
$ \textbf{8.10.2 ::} $
Proportion of adults (15 years and older) with an account at a bank or other financial institution or with a mobile-money-service provider

```python
# 8.1.1 : Annual growth rate of real GDP per capita
water_sanitation_8_1_1 = pd.read_csv('all_indicators/8-1-1.csv')
sns.barplot(data=water_sanitation_8_1_1, x="Year", y="Value")
plt.title('Annual growth rate of real GDP per capital - Rwanda') 
```

```python
# 8.2.1 : Annual growth rate of real GDP per employed person
water_sanitation_8_2_1 = pd.read_csv('all_indicators/8-2-1.csv')
sns.barplot(data=water_sanitation_8_2_1, x="Year", y="Value")
plt.title('Annual growth rate of real GDP per employed person - Rwanda') 
```

```python
# 8.5.1 : Average hourly earnings of female and male employees, by occupation, age and persons with disabilities
water_sanitation_8_5_1 = pd.read_csv('all_indicators/8-5-1.csv')
Observation_level_8_5_1 = pd.pivot_table( water_sanitation_8_5_1,
                                     index = 'Year',
                                    columns = ['Occupation' , 'Sex'],
                                    values = 'Value')
Observation_level_8_5_1.plot(kind='bar')
plt.title('Average hourly earnings - Rwanda') 
plt.xlabel('Year') 
plt.ylabel('Value')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
```

```python
# 8.5.2 : Unemployment rate, by sex, age and persons with disabilities
water_sanitation_8_5_2 = pd.read_csv('all_indicators/8-5-2.csv')
Observation_level_8_5_2 = pd.pivot_table( water_sanitation_8_5_2,
                                     index  = 'Year',
                                     columns  = ['Observation-level footnotes', 'Sex'],
                                    values = 'Value') #print(Observation_level)

Observation_level_8_5_2.plot(kind='bar')
plt.title('Proportion of Unemployment rate, by sex, age and persons with disabilities - Rwanda') 
plt.xlabel('Year') 
plt.ylabel('Value(%)')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left', borderaxespad=0)
```

```python
# 8.10.2 :  Proportion of adults (15 years and older) with an account at a bank or other financial institution or with a mobile-money-service provider
water_sanitation_8_10_2 = pd.read_csv('all_indicators/8-10-2.csv')
sns.barplot(data=water_sanitation_8_10_2, x="Year", y="Value")
plt.title('Proportion of adults (15 years and older) with an account - Rwanda') 
```
